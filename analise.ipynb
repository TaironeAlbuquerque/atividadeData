## Perguntas: Qual a mediana de conceito Enade por estado? E qual a mediana por região do brasil?

### Carregando os dados

```python
import pandas as pd
dados_enade = pd.read_excel('datasete/conceito_enade_2023.xlsx')
dados_enade
```

### Verificando o estado dos dados

```python
dados_enade.info()
```

### Removendo cursos sem notas no Conceito Enade (Contínuo) e adicionando em uma nova variavel para proteger os dados originais

```python
df = dados_enade.dropna(subset=["Conceito Enade (Contínuo)"]).copy()
df
```

### Limpando espaços, agrupando os dados por UF e extraindo a mediana de cada um

```python
df.columns = df.columns.str.strip()
mediana_por_estado = df.groupby("Sigla da UF**")["Conceito Enade (Contínuo)"].median().sort_values(ascending=False)
print(mediana_por_estado)

```

### Função para organizar os dados por região

```python
# Dicionário com os estados por região
regioes = {
    "Norte": ["AC", "AP", "AM", "PA", "RO", "RR", "TO"],
    "Nordeste": ["AL", "BA", "CE", "MA", "PB", "PE", "PI", "RN", "SE"],
    "Centro-Oeste": ["DF", "GO", "MT", "MS"],
    "Sudeste": ["ES", "MG", "RJ", "SP"],
    "Sul": ["PR", "RS", "SC"]
}

# Função que recebe a sigla do estado e retorna a região correspondente
def estado_para_regiao(estado):
    for regiao, estados in regioes.items():
        if estado in estados:
            return regiao
    return "Desconhecido"
```

### Transformando a mediana por estado em um DataFrame, mapeando por região e cauculando a mediana por região

```python
mediana_df = mediana_por_estado.reset_index()
mediana_df["Região"] = mediana_df["Sigla da UF**"].map(estado_para_regiao)
mediana_por_regiao = mediana_df.groupby("Região")["Conceito Enade (Contínuo)"].median().reset_index()
mediana_por_regiao

```

### Verificando os dados dos conceitos por estado antes de transfomar em csv

```python
mediana_por_estado.info()
```

```python
mediana_por_regiao.info()
```

### Gerando o csv da mediana por estado para colocar no BI

```python
mediana_por_estado_df = mediana_por_estado.reset_index()
mediana_por_estado_df.columns = ["UF", "Conceito Enade"]
mediana_por_estado_df.to_csv("datasete/mediana_por_estado.csv", index=False, sep=';', decimal=',')
```
